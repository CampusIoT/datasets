# Balloons :: 2025-05-24 :: IRMA trackers logs



```python
import matplotlib.pyplot as plt
import pandas as pd
```

## Balloon 1

```python
frames = pd.read_csv('./Balloon_1/irma.csv',sep=';',decimal='.')
frames = frames[5000:14000]
#frames = pd.read_csv('./Balloon_1/irma.csv',sep=';',decimal='.',index_col='date')
```

```python
# convert the datetime column to a pandas datetime object
frames['Date_Time'] = pd.to_datetime(frames['date'])

# convert the datetime column to an integer
frames['timestamp'] = frames['Date_Time'].astype(int)

# divide the resulting integer by the number of nanoseconds in milli second
frames['timestamp'] = frames['timestamp'] // 10**6
```

```python
frames
```

```python
temperature = frames["temperature"]
pressure = frames["pressure"]
altitude = frames["altitude"]
vitesse = frames["vitesse"]
vitesse_verticale = frames["vitesse_verticale"]
cog = frames["COG"]
```

```python
temperature.plot()
plt.show()
```

```python
pressure.plot()
plt.show()
```

```python
altitude.plot()
plt.show()
```

```python
vitesse.plot()
plt.show()
```

```python
vitesse_verticale.plot()
plt.show()
```

```python
cog.plot()
plt.show()
```
