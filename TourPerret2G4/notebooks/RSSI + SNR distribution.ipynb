## RSSI Distribution

```python
import pandas as pd
import matplotlib.pyplot as plt

# read in the dataset
df1 = pd.read_csv('../dataset-perret-multichan_gateway_1.csv')
df2 = pd.read_csv('../dataset-perret-multichan_gateway_2.csv')
df3 = pd.read_csv('../dataset-perret-multisf_gateway_2.csv')
```

```python
# draw a histogram of the rssi column
df3['rssi'].hist(bins=20)
df2['rssi'].hist(bins=20)
df1['rssi'].hist(bins=20)

# add labels and title
plt.xlabel('RSSI')
plt.ylabel('Frequency')
plt.title('Distribution of RSSI')
plt.xlim(-130, -80)
```

## SNR Distribution

```python
# read in the dataset
df3 = pd.read_csv('../dataset-perret-multisf_gateway_2.csv')
df2 = pd.read_csv('../dataset-perret-multichan_gateway_2.csv')
df1 = pd.read_csv('../dataset-perret-multichan_gateway_1.csv')
```

```python
# draw a histogram of the snr column
df3['snr'].hist(bins=20)
df1['snr'].hist(bins=20)
df2['snr'].hist(bins=20)

# add labels and title
plt.xlabel('SNR')
plt.ylabel('Frequency')
plt.title('Distribution of SNR')
plt.xlim(-20, 20)
```
